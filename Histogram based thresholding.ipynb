```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv2
```

```python
path = './input/cat4.jpg'
```

```python
def drawImg(img):
    plt.imshow(img,cmap='gray')
    plt.axis("off")
    plt.show()
```

```python
img_ = cv2.imread(path)

img = cv2.cvtColor(img_, cv2.COLOR_BGR2RGB)
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
drawImg(img_gray)
```

```python

# load an image in grayscale mode
img = cv2.imread(path,0)
  
# calculate frequency of pixels in range 0-255
histg = cv2.calcHist([img],[0],None,[256],[0,256]) 
```

```python
drawImg(img)
```

```python
plt.plot(histg)
plt.show()
```

```python
len(histg)
```

```python
a = []

for i in range(0,256):
    x = int(histg[i])
    pair = [i,x]
    a.append(pair)
    
a
```

```python
histArray = np.array(a)
```

## Procedure Th1

```python

def procedure_Threshold1(arrp):
    arrp = arrp[arrp[:,1].argsort()]
    last_pair = arrp[-1]
#     print(last_pair)
    return last_pair[0]
```

```python
th1 = procedure_Threshold1(histArray)
```

```python
th1
```

## Procedure Th2

```python
counter = []
for i in range(0,th1-1):
    counter.append(0)
    
for i in range(0,th1-1):
    for j  in range(i+1,th1):
        fj = int(histg[j])
        fi = int(histg[i])
        if fj<fi:
            counter[i] = counter[i]+1

print(counter)

countpair  = []

for i in range(0,th1-1):
    countpair.append([i,counter[i]])
    
countpair = np.array(countpair)

countpair = countpair[countpair[:,1].argsort()]

countpair
```

```python
countpair

lp = countpair[-1]

th2 = lp[0]
```

```python
th2
```

## Procedure Th

```python
newpairs = []

for i  in range(th2,th1+1):
    zz = int(histg[i])
    pa = [i,zz]
    newpairs.append(pa)
# print(newpairs)
newpairs = np.array(newpairs)

newpairs = newpairs[newpairs[:,1].argsort()]

lpth = newpairs[-1]

th = lpth[0]
```

```python
th
```

```python
def binarize(img1,th):
    H, W = img1.shape
    for r in range(H):
        for c in range(W):
            if img1[r][c]>th:
                img1[r][c]=255
            else:
                img1[r][c]=0
    return img1
```
